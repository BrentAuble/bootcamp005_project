```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from wordcloud import WordCloud
Categories = pd.read_csv('Categories.csv', header=0)
JobList = pd.read_csv('JobsList.csv',header=0)
```

```python
wordcloud = WordCloud(max_font_size=40, relative_scaling=.5).generate(JobList['Src'].to_string())
plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
plt.savefig('wordchart1.png')
```
